```python
import opendatasets as od
od.download('https://www.kaggle.com/datasets/diaaessam/stars-names')
```

```python
import pandas as pd
data = pd.read_csv('/content/stars-names/Stars_Names.csv')
```

```python
data['Approval Date'] = pd.to_datetime(data['Approval Date'])
data['Approval_Year'] = data['Approval Date'].dt.year
data['Approval_Month'] = data['Approval Date'].dt.month
data['Approval_Day'] = data['Approval Date'].dt.day
data.head()
```

```python
df = pd.DataFrame(data)
df
```

```python
print( 'There is', df['Constellation'].nunique(),
      'Constellations and', df['Name'].nunique(), 'Stars')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
Con_bar = df['Constellation'].value_counts()[:63]
Con_bar.plot(kind='bar')
plt.title('The amount of stars in each Constellations')
```

```python
#this code create a dataframe with information of the Zodiac Constellations
Zodiac_Con = df[df['Constellation'].isin(
    ['Scorpius','Virgo','Aries','Taurus','Gemini','Cancer',
     'Leo','Libra','Sagittarius','Capricornus','Aquarius','Pisces'])]
```

```python
Zodiac_Con
```

```python
#this code show the lineplot that shows the number of the stars in each Zodiac constellations
plt.figure(figsize=(10,8))
plt.xlabel('Constellation')
plt.ylabel('Amout_of_Stars')
Zodiac_line = Zodiac_Con['Constellation'].value_counts()
plt.plot(Zodiac_line)
plt.title('The amount of stars in each Constellations')
```

```python
plt.figure(figsize=(12,6))
plt.title('The year and month that stars were approved')
plt.xlabel('Year')
plt.ylabel('Month')
Scorpius_scatter = Zodiac_Con[Zodiac_Con['Constellation'] == 'Scorpius']
Taurus_scatter = Zodiac_Con[Zodiac_Con['Constellation'] == 'Taurus']
plt.scatter(Scorpius_scatter['Approval_Year'], Scorpius_scatter['Approval_Month'])
plt.scatter(Taurus_scatter['Approval_Year'], Taurus_scatter['Approval_Month'])
```

```python
import numpy as np
```

```python
Ap16 = Zodiac_Con[ (Zodiac_Con['Constellation'].isin(['Leo','Libra','Gemini'])) & (Zodiac_Con['Approval_Year'] == 2016)  ]

Leo_num = Ap16[Ap16['Constellation'] == 'Leo']
Libra_num = Ap16[Ap16['Constellation'] == 'Libra']
Gemini_num = Ap16[Ap16['Constellation'] == 'Gemini']
```

```python
Stars_num = np.array([Leo_num['Name'].count(), Libra_num['Name'].count(), Gemini_num['Name'].count()])
Labels = Ap16['Constellation'].unique()
```

```python
plt.title('Amout of Stars in Leo, Libra and Gemini that were approved in 2016')
plt.pie(Stars_num, labels = Labels)
```

```python
plt.title('Amount of Stars in Aries, Aquarius and Cancer that were approved in 2016')
plt.pie(Stars_num, labels = Labels)
```
